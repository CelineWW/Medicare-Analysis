```python
def raw_value(list):
    raw_value_list = []
    for column in list:
        column_split = column.split()
        if column_split[-1] == "value": # and column_split[-2] == "raw"
            raw_value_list.append(column)
        if column_split[-1] != "value":
            continue
    return print(raw_value_list)
raw_value_list = raw_value(column_list)
print(raw_value_list)
```
